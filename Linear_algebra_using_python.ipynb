## The Moor-Penrose Pseudoinverse

```python
Mo = np.array([-1,2,3,-2,5,7]).reshape(3,2)
Mo
```

```python
#SVD
umo, dmo, vmo = np.linalg.svd(Mo)
print(f"left singular vector = \n{umo}\n")
print(f"right singular vector = \n{vmo}\n")
print(f"singular values = {dmo}")
```

```python
# D+
Dmo = np.diag(1/dmo) # aliter: np.linalg.inv(Dmo) snice this is a diagonal matrix so result would be same
Dplus = np.concatenate((Dmo,np.array([[0,0]]).T),axis = 1)
Dplus
```

```python
# calculating A+

Moplus = vmo.T.dot(Dplus.dot(umo.T))
Moplus #pseudoinverse of our matrix
```

```python
#using default function in numpy
Mopl = np.linalg.pinv(Mo)
Mopl
```

### Question

Use torch for the same

```python
Mo_p = torch.tensor(Mo).to(torch.float64)
Mo_p
```

```python
#SVD
umo_p, dmo_p, vmo_p = np.linalg.svd(Mo_p)
print(f"left singular vector = \n{umo_p}\n")
print(f"right singular vector = \n{vmo_p}\n")
print(f"singular values = {dmo_p}")
```

```python
# calculating D+
Dm_p = torch.diag(1/torch.tensor(dmo_p))
zero = torch.tensor(np.array([[0,0]]))
Dm_plus  = torch.concatenate((Dm_p,zero.T),dim = 1)
Dm_plus
```

```python
# pseudoinverse creating
Mopl = torch.tensor(vmo_p).T.matmul(Dm_plus.matmul(torch.tensor(umo_p).T))
Mopl
```

```python
# pseudoinverse creating by default function
Moplus_p = torch.pinverse(Mo_p)
Moplus_p
```

## Regression with pseudoinverse

Using Moor-Penrose Pseudoinverse

For regression problems we typically have many more cases (n, or rows of X) than features to predict (m or columns of X). Let's solve a miniature example of such an overdetermined situationn.

We have n = 8:

```python
xx = np.random.normal(size=(50,))
xx
```

```python
yy = np.random.normal(size=(50,))
yy
```

```python
fig, ax = plt.subplots()
plt.title("Clinical trails")
plt.xlabel("Drug dosage (ml)")
plt.ylabel("Forgetfulness score")
ax.scatter(xx,yy)
```

```python
x0 = np.ones(50)
x0
```

```python
xxx = np.concatenate((np.matrix(x0).T,np.matrix(xx).T),axis=1)
xxx
```

```python
w = np.dot(np.linalg.pinv(xxx),yy)
w
```

```python
b = np.asarray(w).reshape(-1)[0]
b
```

```python
m = np.asarray(w).reshape(-1)[1]
m
```

```python
fig, ax = plt.subplots()
plt.title("Clinical trails")
plt.xlabel("Drug dosage (ml)")
plt.ylabel("Forgetfulness score")
ax.scatter(xx,yy)
x_min,x_max = ax.get_xlim()
y_min,y_max = m*x_min+b , m*x_max+b

ax.set_xlim([x_min,x_max])
ax.plot([x_min,x_max],[y_min,y_max],"--k")
```
